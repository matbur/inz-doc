```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y = 1 / (1 + np.exp(-x))

plt.title('Sigmoid')
plt.plot(x, y, label ='sigmoid')
plt.legend(loc='upper left')
```

```python
y = 2 / (1 + np.exp(-2 * x)) - 1

plt.title('Tangens hiperboliczny')
plt.plot(x, y, label ='tanh')
plt.legend(loc='upper left')
```
